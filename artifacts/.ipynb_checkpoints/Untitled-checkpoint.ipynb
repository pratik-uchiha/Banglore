```python
def get_house_price(location,sqft,bath,bhk): 

    test_array = np.zeros(model.n_features_in_,int)
    index = column_names.index(location)
    test_array[0] = sqft
    test_array[1] = bath
    test_array[2] = bhk
    test_array[index]= 1

    price = np.around(model.predict([test_array])[0],3)
    return price
```

```python
get_house_price('amruthahalli',1300,4,4)
```
